#### Find The Skewness of All The Columns

```python
# Print out the skewness of each features

print('Skewness of all the features: ')
skew = fdata.skew()
skew = skew.sort_values()
print(skew)
```

#### Find The Correlation Values For All Columns

```python
# Returns the correlation matrix of the dataframe
fdata.corr()
```

#### Graphical Representation of Correlations

```python
# Draw the heatmap with the mask and correct aspect ratio
corr = fdata.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# visualize the correlation of all columns in this dataframe
plt.figure(figsize=(15,15))

sns.heatmap(corr, mask=mask, vmax=.3, 
            square=True, linewidths=.5, cbar_kws={"shrink": .3})
```

#### Remove any cols that starts with 'Soil'

```python
# get and store all the not Soil Type features
columns = [cols for cols in fdata.columns if not str(cols).startswith('Soil')]

# set the figure size
plt.figure(figsize=(9, 9))

# using heapmap to plot
sns.heatmap(fdata[columns].corr(), square=True, linewidths=1, annot = True, fmt='.0%', cbar_kws={"shrink": .3}, vmax=.3)

# set the title of the plot
plt.title('Correlation of the features')

# show the plot
plt.show()
```

### Cover_Type & The Different Wilderness Visualizations

```python
# define the columns that want to plot out
w_area = ['Wilderness_Area_1', 'Wilderness_Area_2', 'Wilderness_Area_3', 'Wilderness_Area_4']
    
# using for loop to generate the graph
for i in range(len(w_area)):
    
    # plot i number of graphs
    plt.figure(i, figsize=(5, 3.5))

    # group by each of the cover type and sum each of them
    fdata.groupby('Cover_Type')[w_area[i]].sum().plot.bar()
    
    # display the title name
    plt.title(w_area[i])
```
